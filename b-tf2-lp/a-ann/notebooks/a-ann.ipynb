### Create the Dataset

```python
import pandas as pd
import numpy as np

# Define the number of samples
num_samples = 30000

# Generate sensor data in the range [0, 100]
sensor_data = np.random.randint(0, 101, size=(num_samples, 8))

# Create a DataFrame for sensor data
sensor_df = pd.DataFrame(sensor_data, columns=['sensor1', 'sensor2', 'sensor3', 'sensor4', 'sensor5', 'sensor6', 'sensor7', 'sensor8'])

sensor_df.head()
```

```python
def calculate_throttle(sensor_data):
    avg_sensor_value = np.sqrt(np.mean(sensor_data))
    throttle = avg_sensor_value / 10  # Adjust scale if needed
    return throttle

def calculate_steering_angle(row):
    sum_first_four_sensors = row[['sensor1', 'sensor2', 'sensor3', 'sensor4']].sum()
    prod_last_four_sensors = row[['sensor5', 'sensor6', 'sensor7', 'sensor8']].prod()
    steering_angle = min(60, (prod_last_four_sensors/sum_first_four_sensors - 200) / 1000)  # Adjust scale if needed
    return steering_angle
```

```python
# Apply functions to calculate throttle and steering angle
sensor_df['throttle'] = sensor_df.apply(calculate_throttle, axis=1)
sensor_df['steering_angle'] = sensor_df.apply(calculate_steering_angle, axis=1)

# Display the first few rows of the dataset
sensor_df.head()
```
